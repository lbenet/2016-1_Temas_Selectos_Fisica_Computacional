###Ejercicio 2: Integración con Runge–Kutta

$\dot{x}(t)=x^2$, $x(0)=1$

Haciendo separacion de variables, la solución analítica con la condición inicial $x(0)=1$ es:
$$x=\dfrac{1}{1-t}$$

Una ilustración de la solución se puede ver aquí
![](http://img-cdn.jg.jugem.jp/881/1922510/20101201_1467921.jpg)
$$Imagen 1$$

```julia
function rk4(f,h,n,x0,t0)
    x_actual=x0
    t_actual=t0
    xs = zeros(n)
    ts = zeros(n)
    for i in 1:n
        k1 = f(t_actual,x_actual)*h
        k2 = f(t_actual + h/2, x_actual + k1/2)*h
        k3 = f(t_actual +h/2, x_actual + k2/2)*h
        k4 = f(t_actual + h, x_actual + k3)*h
        
        k = (k1 + 2k2+ 2k3 + k4)/6
        
        t_nueva = t_actual + h
        x_nueva = x_actual + k
        
        t_actual = t_nueva
        x_actual = x_nueva
        
        xs[i]= x_actual
        ts[i]= t_actual
    end
    ts,xs
end
```

```julia
f(t,x)=x^2
```

```julia
rk4(f,0.0625, 15, 1, 0)
```

```julia
using PyPlot
```

```julia
sol_analitica(t) = 1/(1-t)
```

```julia
xa = [0:0.0625:1-0.0625]
ya = [1/(1-t) for t in xa]

plot(xa,ya,"r h")
```

```julia
xrk4,yrk4 = rk4(f,0.0625, 15, 1, 0)

plot(xrk4,yrk4,"g h")
```

Para poder comparar las dos soluciones (analitica y Runge-Kutta) las graficaremos juntas

```julia
plot(xa,ya,"r .")
plot(xrk4,yrk4,"g")
```
